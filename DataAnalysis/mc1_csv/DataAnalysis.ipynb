```python
#Merging DataFrame

import pandas as pd

# Load your node and edge CSV files
nodes_df = pd.read_csv('mc1_nodes.csv')

edges_df = pd.read_csv('mc1_edges.csv')

# Merging the data on 'id' from nodes_df and 'source' from edges_df
merged_df = pd.merge(edges_df, nodes_df, left_on='source', right_on='id', how='left')


# Displaying the merged dataframe
merged_df
```

```python
# Filter the rows where 'notable' is False
notable_false_df = merged_df[merged_df['notable'] == False]

# Display the filtered dataframe
notable_false_df
```

```python
# Filter the rows where 'notoriety_date' is not the same as 'release_date'
filtered_df = merged_df[merged_df['notoriety_date'] != merged_df['release_date']]

# Display the filtered dataframe
filtered_df
```

```python
# Filter the rows where 'notoriety_date' is not empty (not NaN)
filtered_non_empty_notoriety = filtered_df[filtered_df['notoriety_date'].notna()]

# Display the filtered dataframe
filtered_non_empty_notoriety
```

```python
# Dropping the 'source' and 'target' columns from the merged dataframe

import pandas as pd

merged_df_cleaned = merged_df.drop(columns=['source', 'target','key','id','stage_name'])
merged_df_cleaned.head(5)
```

```python
#Fill Values from before until the next value changes
columns_to_fill = ['single', 'release_date', 'notable','genre']
merged_df_cleaned[columns_to_fill] = merged_df_cleaned[columns_to_fill].fillna(method='ffill')
#merged_df_cleaned.head(50)
merged_df_cleaned
```

```python
# Apply the condition: Replace NaN in 'notoriety_date' with 'release_date' if 'release_date' >= 'notoriety_date'
merged_df_cleaned['notoriety_date'] = merged_df_cleaned.apply(
    lambda row: row['release_date'] if pd.isna(row['notoriety_date']) and row['release_date'] >= row['release_date'] else row['notoriety_date'],
    axis=1
)

# Display the first 50 rows of the cleaned dataframe
#merged_df_cleaned.head(50)
merged_df_cleaned
```

```python
# Replace NaN values in 'written_date' with the corresponding value from 'release_date'
merged_df_cleaned['written_date'] = merged_df_cleaned['written_date'].fillna(merged_df_cleaned['release_date'])

# Display the modified dataframe
merged_df_cleaned.head(50)
#merged_df_cleaned
```

```python
# Convert the 'release_date', 'written_date', and 'notoriety_date' columns to integer year
merged_df_cleaned['release_date'] = merged_df_cleaned['release_date'].astype('Int64')
merged_df_cleaned['written_date'] = merged_df_cleaned['written_date'].astype('Int64')
merged_df_cleaned['notoriety_date'] = merged_df_cleaned['notoriety_date'].astype('Int64')

# Display the updated dataframe
merged_df_cleaned.head(50)
```
